```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Kernel_Regression:
    
    def _gaussian_kernel(self, x1, x2, var):
        return math.e**(-(x1 - x2)**2 / (2 * var**2))
    
    def fit(self, X, y):
        self.X = X
        self.y = y
        
    def predict(self, x, var=2):
        k = self._gaussian_kernel(x, self.X, var)
        return k.T @ self.y / sum(k)
```

```python
X = np.arange(0, 100, 0.5)
y = np.sin(X) + np.random.randn(X.shape[0]) + np.cos(X)

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
L = Kernel_Regression()
L.fit(X_train, y_train)
y_pred = []
for i in X_test:
    y_pred.append(L.predict(i, 1))
```

```python
plt.plot(y_test)
plt.plot(y_pred)
plt.show()
```

```python
mean_squared_error(y_test, y_pred)
```
