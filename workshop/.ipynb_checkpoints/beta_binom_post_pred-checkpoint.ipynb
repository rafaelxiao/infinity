```python
def binomial_beta_plot(params):
    x = np.linspace(0.001, 0.999, 50)
    plt.figure(figsize=(8, 6))
    for i in params:
        a = i[0]
        b = i[1]
        n0 = i[2]
        n1 = i[3]
        text = "a:%s, b:%s, n0:%s, n1:%s"%(a, b, n0, n1)
        plt.plot(np.exp(beta_log_prob(x, a+n0, b+n1)), label=text)
    plt.legend()
    plt.show()
```

```python
# Prior, Likelihood, Posterior
binomial_beta_plot([
    [2, 2, 0, 0],
    [1, 1, 4, 1],
    [2, 2, 4, 1],
])
```
